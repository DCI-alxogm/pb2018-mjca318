```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h=0.00033
theta=np.zeros(1000)
theta[0]=10
i=np.arange(0,1000,h)
w=20
v=np.zeros(1000)
```

```python
for i in range(0,999,1):
    theta[i+1]=theta[i]+h*v[i]
    v[i+1]=v[i]+h*(-w**2*theta[i])
```

```python
plt.plot(v,'blue')
plt.xlabel("v")
```

```python
plt.plot(theta,'purple')
plt.xlabel('theta')
```

```python
plt.plot(v,theta,'orange')
plt.xlabel("v")
plt.ylabel("theta")
```
